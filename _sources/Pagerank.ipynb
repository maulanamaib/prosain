## Hitung Manual


Rumus yang saya gunakan disini adalah:
PR(A) = (1 - d) / N + d * sum(PR(j) / L(j))
dengan:
n = 0,85
N = 9

*   PR(A) adalah nilai PageRank dari node A
Item daftar

*   d adalah dampening factor yang biasanya diatur pada nilai 0.85
Item daftar

*   N adalah jumlah total node pada graph
Item daftar
*   PR(j) adalah nilai PageRank dari setiap node Ti yang terhubung langsung ke node A
Item daftar

*   L(j) adalah jumlah edge keluar dari setiap node Ti yang terhubung langsung ke node AItem daftar

* Node 2: 2
