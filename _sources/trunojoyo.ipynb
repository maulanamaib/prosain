```python
from requests_html import HTMLSession 
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
s = HTMLSession()
```

```python
def getData(url):
  r = s.get(url)
  soup = bs(r.text,"html.parser")
  return soup

```

```python
def getNextPage(url):
  page = url.find("ol", {"class" : "pagination"})
  link = page.find(["a"], class_ = "pag_button", text = ">")
  if not link:
    return
  else:
    url =(f"{link['href']}")
    return url
```

```python
url = "https://pta.trunojoyo.ac.id/c_search/byprod/10"
```

```python
Go = True
hasil = {"judul": [], "abstrak" : []}
while Go:
  soup_prodi = getData(url)
  all_links = soup_prodi.find_all(["a"], class_ = "gray button")
  for link in all_links:
    link_jurnal = str(f"{link['href']}")
    soup_jurnal = getData(link_jurnal)
    title = soup_jurnal.find(["a"], class_ = "title")
    # print(title.string)
    hasil["judul"].append(title.string)
    abstrak = soup_jurnal.find(["p"], align = "justify")
    hasil["abstrak"].append(abstrak.string)
    # print(abstrak.string)
  soup = getData(url)
  url = getNextPage(soup)
  if not url: 
    Go = False
```

```python
df = pd.DataFrame(list(zip(hasil["judul"],hasil["abstrak"])),columns = ["Judul","Abstrak"])
```
